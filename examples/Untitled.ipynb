```python
r = 0.9
n = 10
cov = np.array([[1, r], [r, 1]])
cov = np.ones((n, n)) * r
cov[np.diag_indices(n)] = 1

t = np.arange(0, 1000, 1)

ou = MultivariateOUProcess(mu=0, cov=cov, tau=10)
x = ou.sample(t, shape=(cov.shape[0],))

fig, ax = plt.subplots()
ax.plot(t, x[:, :3])
```
